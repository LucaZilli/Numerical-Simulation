```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ave_g_liquid = np.loadtxt(fname = "output.gave.0",usecols=0)
err_g_liquid = np.loadtxt(fname = "output.gave.0",usecols=1)
```

```python
npart=108.0
rho=0.8
vol=npart/rho
box=vol**(1./3.)
L=box/2.
bins = np.linspace(start=0, stop=L, num=100)
#print(bins)
#print(ave_g_liquid)
bins=np.asarray(bins)
ave_g_liquid=np.asarray(ave_g_liquid)
err_g_liquid=np.asarray(err_g_liquid)

#plt.style.use('seaborn-whitegrid')
plt.errorbar(bins,ave_g_liquid,yerr=err_g_liquid,lw=1,color='black',
             ecolor='black', elinewidth=5)
#plt.plot(bins,ave_g_liquid,lw=1)
plt.xlabel('r')
plt.ylabel('g(r)')
plt.grid(True)
plt.show()
```
