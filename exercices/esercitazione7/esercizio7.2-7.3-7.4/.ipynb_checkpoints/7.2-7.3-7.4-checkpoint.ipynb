### EXERCISE 07.2-07.3-07.4


From the analysis of the excercise 7.1 it is known that the minimum number of steps per block is 1000 for every phase. Indeed for the simulations have been chosen blocks large 5000 steps to be sure that they are large enough without falling in efficiency problems (i.e. simulations too long).

Indeed, after added in the code the sampling of the radial distribution function, It has been simulated the Argon element in each phase with the following input data:
1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

Subsequently have been reported the measurements of the internal energy per particles, pressure and radial distribution as function of blocks.

Every measurements have been taken after having equilibrated the system. To be sure the system was equilibrated it has been looked at the temperature: after at least an oscillation around the input value it can be sure that the system has reached the configuration of the equilibrium.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Lennard-Jones units:

```python
sigma=0.34e-9
Kb=1.380649e-23
epsilon=120*Kb
```

### SOLID PHASE:

In these python cells have been plotted the radial distribution, the energy per particles and the pression in SI units as functions of number of blocks for the **solid** phase. The errors have been estimated as the standard deviation of the mean.

```python
epot_solid = np.loadtxt(fname = "ave_Epot_solid.txt",usecols=2)
err_epot_solid = np.loadtxt(fname = "ave_Epot_solid.txt",usecols=3)
epot_solid *= epsilon
err_epot_solid *= epsilon

g_solid = np.loadtxt(fname = "ave_G_solid.txt",usecols=0)
err_g_solid = np.loadtxt(fname = "ave_G_solid.txt",usecols=1)

g_verlet_solid = np.loadtxt(fname = "ave_G_solid_verlet.txt")
err_g_verlet_solid = np.loadtxt(fname = "err_G_solid_verlet.txt")



pres_solid = np.loadtxt(fname = "ave_Pres_solid.txt",usecols=2)
err_pres_solid = np.loadtxt(fname = "ave_Pres_solid.txt",usecols=3)
pres_solid *= (epsilon/sigma**3)
err_pres_solid *= (epsilon/sigma**3)
```

```python
N=100
t = np.arange(N)      # [0,1,2,...,N-1]

"""Plots and layout"""
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(17, 12))

"""Potential Enrgy"""
ax1.errorbar(t,epot_solid,yerr=err_epot_solid,color='blue')
ax1.set_title(f'Potential Energy',  fontsize=22);
ax1.set_xlabel("# Blocks",fontsize=20)
ax1.set_ylabel("U/N [J]",fontsize=20)
ax1.tick_params(axis="x", labelsize=15)
ax1.tick_params(axis="y", labelsize=15)
ax1.grid(True)

"""Pression"""
ax2.errorbar(t,pres_solid,yerr=err_pres_solid,color='green')
ax2.set_title(f'Pression',  fontsize=22);
ax2.set_xlabel("# Blocks",fontsize=20)
ax2.set_ylabel("P [Pa]",fontsize=20)
ax2.tick_params(axis="x", labelsize=15)
ax2.tick_params(axis="y", labelsize=15)
ax2.grid(True)



"""g(r)"""
npart=108.0
rho=1.1
vol=npart/rho
box=vol**(1./3.)
L=box/2.*sigma
bins = np.linspace(start=0, stop=L, num=100)

"""g(r) Monte Carlo"""
ax3.errorbar(bins,g_solid,yerr=err_g_solid,
             lw=1,
             color='black',
             ecolor='black', 
             elinewidth=5)
ax3.set_title(f'g(r): Monte Carlo Simulation',  fontsize=22);
ax3.set_xlabel("r",fontsize=20)
ax3.set_ylabel("g(r)",fontsize=20)
ax3.tick_params(axis="x", labelsize=15)
ax3.tick_params(axis="y", labelsize=15)
ax3.grid(True)

"""g(r) Verlet algorithm"""
bins = np.linspace(start=0, stop=L, num=100)
ax4.errorbar(bins,g_verlet_solid,yerr=err_g_verlet_solid,
             lw=1,
             color='black',
             ecolor='black', 
             elinewidth=5)
ax4.set_title(f'g(r): Verlet Algorithm Simulation',  fontsize=22);
ax4.set_xlabel("r [m]",fontsize=20)
ax4.set_ylabel("g(r)",fontsize=20)
ax4.tick_params(axis="x", labelsize=15)
ax4.tick_params(axis="y", labelsize=15)
ax4.grid(True)

fig.suptitle(f'Potential Energy, Pression and g(r) of the $Solid$ Phase:\n averages and error as function of number of blocks ',  fontsize=25);
fig.tight_layout()
fig.subplots_adjust(top=0.8,hspace = 0.5,wspace = 0.4)
plt.show()
```

To verify if the distribution of g(r) sampled with two diferrent algorithm of evolution is consistent it has been plotted the distribution of differences in the cells below.

```python
plt.errorbar(bins,g_verlet_solid-g_solid,yerr=err_g_verlet_solid+err_g_verlet_solid,label="differences",c="green")
plt.title(f'g(r): Verlet Algorithm - Monte Carlo Simulation',  fontsize=22);
plt.xlabel("r [m]",fontsize=20)
plt.ylabel("g(r)",fontsize=20)
plt.tick_params(axis="x", labelsize=15)
plt.tick_params(axis="y", labelsize=15)
plt.hlines(0, 0, 8*10**(-10), label='Theory')
plt.grid(True)
leg = plt.legend(loc=(0.1, 0.1),frameon=True)
leg.get_frame().set_edgecolor('black')
leg.get_frame().set_linewidth(3)
```
