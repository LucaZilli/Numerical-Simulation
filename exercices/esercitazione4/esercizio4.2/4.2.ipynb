### EXERCISE 04.2

In this exercise has been implemented the subdivions in blocks of the data measured. In this jupyter notebook has been explained how this implementation has been done and has been plotted an example of the output with relatives comments.

After an Initiale phase where the systems must be equilibrated as explained In the excercise 4.1, the measurements can be taken.
In the simulations below $10^5$ steps have been made and the measurements have been taken every 10 steps, Indeed has been made an analysis with $10^4$ values divided in 100 blocks.


On the c++ code all the measurements have been stored in vectors "data_namequantity". At the end of the simulation, in the function "main" the data have been divided in 100 blocks and subsequently printed in files.

Files named "ave_namequantity.out" contain the data measured with blocking methods and "errori_namequantity.out" contain relatives errors.
Files named "data_namequantity.out", instead, contain all the measurements taken during the simulations

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
M=100000
N=100
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
L=int(M/(N*10))
x = np.arange(N) 


#qual è l'unità di misura dell'energia?


"""potential energy"""
sum_prog_epot = np.loadtxt(fname = "ave_epot.out")
err_prog_epot = np.loadtxt(fname = "errori_epot.out")
      
x*=L # Number of experiments = block * (Number of experiments in each block)
plt.errorbar(x,sum_prog_epot,yerr=err_prog_epot)
plt.xlabel('#measures')
plt.ylabel('epot')
plt.grid(True)
plt.show()

"""kinetic energy"""
sum_prog_ekin = np.loadtxt(fname = "ave_ekin.out")
err_prog_ekin = np.loadtxt(fname = "errori_ekin.out")
      
plt.errorbar(x,sum_prog_ekin,yerr=err_prog_ekin)
plt.xlabel('#measures')
plt.ylabel('ekin')
plt.grid(True)
plt.show()

"""total energy"""
sum_prog_etot = np.loadtxt(fname = "ave_etot.out")
err_prog_etot = np.loadtxt(fname = "errori_etot.out")
      

plt.errorbar(x,sum_prog_etot,yerr=err_prog_etot)
plt.xlabel('#measures')
plt.ylabel('etot')
plt.grid(True)
plt.show()

"""temperature"""
sum_prog_temp = np.loadtxt(fname = "ave_temp.out")
err_prog_temp = np.loadtxt(fname = "errori_temp.out")
      

plt.errorbar(x,sum_prog_temp,yerr=err_prog_temp)
plt.xlabel('#measures')
plt.ylabel('temp')
plt.grid(True)

plt.show()
```

```python
plt.figure()
epot = np.loadtxt(fname = "data_epot.out")
ekin = np.loadtxt(fname = "data_ekin.out")
etot = np.loadtxt(fname = "data_etot.out")
temp = np.loadtxt(fname = "data_temp.out")

x=np.arange(epot.size)
plt.plot(x,epot, label='potential energy')
plt.plot(x,ekin, label='kinetic energy')
plt.plot(x,etot, label='total energy')
plt.plot(x,temp, label='temperature')
plt.xlabel('number of measures')
plt.grid(True)
plt.title(f'Measures of the simulation\ndoes the Total Energy fluctuate?')
plt.legend(loc=(0.60,0.47))

plt.show()
```

The simulation is implemented in the microcanonical ensamble. Indeed the time evolution is controlled by the *Verlet algorithm*, that integrate the dynamical equation until the second order like 
$$
\vec{r}(t+dt)\approx2\vec{r}(t)−\vec{r}(t−dt)+\frac{dt^2}{m}\vec{F}(t)
$$

The choice is not casual, because a peculiarity of this algorithm is to be time reversable that causes the conservation of the Total Energy like must be in microcanonical ensambles.

This can be seen from the last plot, where all the data have been printed on the same plane and is shown very clear that all the quantities have big fluctuations around their mean value, rather the Total Energy that is conserved on the time evolution.
