### EXERCISE 04.1

In this exercise has been simulated a systems of many particles interacting with a potential Lennard-Jones.
The Molecular Dynamics (MD) code starts the simulation run from a provided spatial configuration of the $N$ particles (file: <font face="Courier">config.0</font>) and, once the simulation is finished, it writes out the final configuration (file: <font face="Courier">config.final</font>). 


### Modifications of the c++ code:

It has been added a new variable **m_BOOL** to the code and it can be modified from the file "*input.dat*". When **m_BOOL=0** the algorithm reads the initial configuration from the file "*config.0*" and choose the inital velocities randomly. When **m_BOOL=1** the algorithm starts the simulation reading from the final configuration of the last simulation made and sets the velocities as the mean of the positions of the last configuration and the positions of the second last configuration of the previous simulation made.

Indeed, if someone wants to equilibrate the system must set **m_BOOL=1** and start the new simulation after having copied the last configuration and the second last configuration respectively from the file "config.final" and "old.final" to the file "config.0" and "old.0".

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
output_temp = np.loadtxt(fname = "output_temp.dat")

x = np.arange(output_temp.size)

plt.errorbar(x,output_temp)
plt.xlabel('measures')
plt.ylabel('T')
plt.grid(True)
plt.show()
```

The iamge above is a graph of an evolution started from a **fcc configuration**.
As it can be seen from the plot upon, every time a simulation is started the temperature fall down due to the potential energy, which at the initial configuration, is strong and tend to minimize the movements of the atoms. 

After 4 simulations the system can be considered at the equilibrium: has reached the desired temperature and fluctuates around it.
