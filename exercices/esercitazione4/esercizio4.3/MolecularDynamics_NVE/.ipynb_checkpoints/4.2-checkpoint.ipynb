```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
M=100000
N=100
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
L=int(M/(N*10))
x = np.arange(N) 


#qual  l'unit di misura dell'energia?


"""potential energy"""
sum_prog = np.loadtxt(fname = "ave_epot.out")
err_prog = np.loadtxt(fname = "errori_epot.out")
      
x*=L # Number of experiments = block * (Number of experiments in each block)
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel('#measures')
plt.ylabel('epot')
plt.grid(True)
plt.show()

"""kinetic energy"""
sum_prog = np.loadtxt(fname = "ave_ekin.out")
err_prog = np.loadtxt(fname = "errori_ekin.out")
      
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel('#measures')
plt.ylabel('ekin')
plt.grid(True)
plt.show()

"""total energy"""
sum_prog = np.loadtxt(fname = "ave_etot.out")
err_prog = np.loadtxt(fname = "errori_etot.out")
      

plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel('#measures')
plt.ylabel('etot')
plt.grid(True)
plt.show()

"""temperature"""
sum_prog = np.loadtxt(fname = "ave_temp.out")
err_prog = np.loadtxt(fname = "errori_temp.out")
      

plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel('#measures')
plt.ylabel('temp [k]')
plt.grid(True)
plt.show()
```
