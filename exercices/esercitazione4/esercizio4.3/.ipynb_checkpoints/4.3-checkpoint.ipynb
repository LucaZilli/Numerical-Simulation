### EXERCISE 04.3

In this exercise has been implemented a simulation of the element argon in every phase: solid, liquid, gas. Eeach phase of matter has been implemented setting the input parameters as follows:
1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

After the data of the measurements have been collected, have been shown pictures of the estimations of the quantities of the systems in SI units as function of the blocks.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
M=100000
N=100
L=int(M/(N*10))
x = np.arange(N) 
x*=L # Number of experiments = block * (Number of experiments in each block)
```

For this excercise the simulation has been carry on as explained in excercise 4.2 just changing the "input.dat" file given with the values of the solid phase of Argon shown above.
After at least an oscillation of the temperature around the value fixed in the "input.dat" file, the system is considered equilibrated.

During the simulation have been taken $10^5$ steps with $10^4$ measurements divided in 100 blocks and here below have been plotted the measurements with relatives errors as function of blocks. The measurements have been changed in SI units directly in the c++ code, after having verified in LJ units that the results obtained make sense. Indeed in LJ units the quantities are expressed on the order of $10^1$ and the research of some errors is more effective.

### Solid:

#### Equilibration:

```python
output_temp = np.loadtxt(fname = "output_temp_solid.dat")

x_eq = np.arange(output_temp.size)

plt.errorbar(x_eq,output_temp)
plt.xlabel('# measures')
plt.ylabel(f'$T^\star$')
plt.grid(True)
plt.show()

output_epot = np.loadtxt(fname = "output_epot_solid.dat")

plt.errorbar(x_eq,output_epot)
plt.xlabel('# measures')
plt.ylabel(f'$E_P^\star$')
plt.grid(True)
plt.show()
```

### Gas

#### Equilibration

```python
output_temp = np.loadtxt(fname = "output_temp_gas.dat")

x_eq = np.arange(output_temp.size)

plt.errorbar(x_eq,output_temp)
plt.xlabel('# measures')
plt.ylabel(f'$T^\star$')
plt.grid(True)
plt.show()

output_epot = np.loadtxt(fname = "output_epot_gas.dat")

plt.errorbar(x_eq,output_epot)
plt.xlabel('# measures')
plt.ylabel(f'$E_P^\star$')
plt.grid(True)
plt.show()
```

##### An observation:

In the equilibration of the solid and the liquid phase, the system must rise up the temperature near the target value. On the other hand, in the simulation of the gas phase happens the opposite. This can be explained by the density parameter imposted in the "input.dat" file. 
For the liquid and solid phase the denisty is, more or less, an order of magnitude more than the gas phase, and when the simulation is started the particles are near to each other and the potential of Lennard-Jones dominates over the kinetic energy (i.e. over the temperture). In the simulation of the gas phase the particles are so far that the situation is reversed and the potential is dominated by the kinetic term. This situation brings the temperature to a minor value to that imposed by the input file.
