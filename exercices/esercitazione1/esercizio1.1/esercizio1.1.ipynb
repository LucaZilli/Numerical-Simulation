```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
M=100000
N=100
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
L=int(M/N)
x = np.arange(N) 


sum_prog = np.loadtxt(fname = "valori_medie.txt")
err_prog = np.loadtxt(fname = "errori_medie.txt")
      
      

x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,sum_prog-0.5,yerr=err_prog)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

M=100
dati = np.zeros(100000)
dati= np.loadtxt(fname = "dati.txt")
n_s=np.zeros(100)
n_t=1000/M
chi=np.zeros(100)
x=np.zeros(100)
err = np.zeros(M)

for j in range(M):
    x[j]=j+1

for i in range(100):#in questo ciclo riempo i chi quadro
    sper=np.zeros(1000)
    for j in range(1000):#a ogni ciclo di chi quadro traslo i dati di 1000
        prog=i*1000+j
        sper[j]=dati[prog]
    
    for j in range(100):#conto i valori distribuiti dentro ogni cella che dovrebbero essere 10
        for k in range(1000):
            if (sper[k]<10*(j+1)):
                n_s[j]=n_s[j]+1
            
    
    for j in range(M):#faccio il chi quadro
        chi[i]+=((n_s[j]-n_t)**2)/n_t

    err[i]=n_t
    
for j in range(M):
    print(chi[i])
    
plt.errorbar(x,chi,yerr=err)
plt.xlabel('n')
plt.ylabel('chi quadro')
plt.grid(True)
plt.show()    
```
