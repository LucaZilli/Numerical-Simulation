## Exercise 01.2
In this excercise has been extended the Pseudo-Random Number generator downloaded from the NSL Ariel web site and has been checked the Central Limit Theorem:

Have been added to the class two methods that provide numbers distributed as the Cauchy and as the Exponential distributions.

To control if the data sampled satisfy the central limit theorem has been implemented a code that returns 12 files. The files has been divided in three class. Each class is composted by 4 files filled with $N=10^4$ data of function $S_1$, $S_2$, $S_{10}$ and $S_{100}$, where the function $S_i$ is an average of i values that follow a particular distribution.

1. For the first class has been used the uniform distribution in the interval $[0,1]$.

2. For the second class has been used the exponential distribution with mean equal to one $p(x)=e^{-x}$ in the interval $[0,\infty)$.

3. For the third class has been used the cauchy distribution $p(x)=\frac{1}{\pi}\frac{\gamma}{(x-x_0)^2+\gamma^2}$ with $x_0=0$ and $\gamma=1$.

If the central limit theorem is really true and the code has been implemented correctly, the means of data distributed uniformly and the means data distributed exponentialy must converge to a gaussian distribution with mean eaqual to the mean of the original distribution ($\mu=\mu_0$) and variance equal to the original variance on the number used to calculate the average ($\sigma^2=\sigma_0^2/N$). Instead the function $S_i$ of the data Cauchy-distributed must converge to another Cauchy distribution.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def gaussian(x, mu, sig):
    return 1./(np.sqrt(2.*np.pi)*sig)*np.exp(-np.power((x - mu)/sig, 2.)/2)

def cauchy(x, x0, gamma):
    return 1./np.pi*gamma/((x-x0)**2+gamma**2)
```

```python
Data_Cauchy_S1 = np.zeros(10000)
Data_Cauchy_S1 = np.loadtxt(fname = "Data_Cauchy_S1.txt")

Data_Cauchy_S2 = np.zeros(10000)
Data_Cauchy_S2 = np.loadtxt(fname = "Data_Cauchy_S2.txt")

Data_Cauchy_S10 = np.zeros(10000)
Data_Cauchy_S10 = np.loadtxt(fname = "Data_Cauchy_S10.txt")

Data_Cauchy_S100 = np.zeros(10000)
Data_Cauchy_S100 = np.loadtxt(fname = "Data_Cauchy_S100.txt")


Data_Exp_S1 = np.zeros(10000)
Data_Exp_S1 = np.loadtxt(fname = "Data_Exp_S1.txt")

Data_Exp_S2 = np.zeros(10000)
Data_Exp_S2 = np.loadtxt(fname = "Data_Exp_S2.txt")

Data_Exp_S10 = np.zeros(10000)
Data_Exp_S10 = np.loadtxt(fname = "Data_Exp_S10.txt")

Data_Exp_S100 = np.zeros(10000)
Data_Exp_S100 = np.loadtxt(fname = "Data_Exp_S100.txt")


Data_Unif_S1 = np.zeros(10000)
Data_Unif_S1 = np.loadtxt(fname = "Data_Unif_S1.txt")

Data_Unif_S2 = np.zeros(10000)
Data_Unif_S2 = np.loadtxt(fname = "Data_Unif_S2.txt")

Data_Unif_S10 = np.zeros(10000)
Data_Unif_S10 = np.loadtxt(fname = "Data_Unif_S10.txt")

Data_Unif_S100 = np.zeros(10000)
Data_Unif_S100 = np.loadtxt(fname = "Data_Unif_S100.txt")
```

### Standard dice (i.e. uniform distribution)

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(17, 12))

ax1.hist(Data_Unif_S1,bins=100,range=(-0.5,1.5), color="aquamarine",density=True)
ax1.set_xlabel('X[Unif]',fontsize=20)
ax1.set_ylabel('NUMBERS',fontsize=20)
ax1.grid(True)
ax1.set_title(f'S1',  fontsize=22);

ax2.hist(Data_Unif_S2,bins=100,range=(-0.5,1.5), color="aqua",density=True)
ax2.set_xlabel('S2[Unif]',fontsize=20)
ax2.set_ylabel('NUMBERS',fontsize=20)
ax2.grid(True)
ax2.set_title(f'S2',  fontsize=22);

ax3.hist(Data_Unif_S10,bins=100,range=(0.2,0.8), color="deepskyblue",density=True)
ax3.set_xlabel('S10[Unif]',fontsize=20)
ax3.set_ylabel('NUMBERS',fontsize=20)
ax3.grid(True)
ax3.set_title(f'S10',  fontsize=22);

Bins=np.linspace(0.4,0.6,100)
gauss=np.array([gaussian(x,0.5,np.sqrt(1./1200)) for x in Bins])
ax4.hist(Data_Unif_S100,bins=100,range=(0.4,0.6), color="blue",density=True,label="numerical")
ax4.plot(Bins,gauss, color="red",label="theorical")
ax4.set_xlabel('S100[Unif]',fontsize=20)
ax4.set_ylabel('NUMBERS',fontsize=20)
ax4.grid(True)
ax4.legend(fontsize=20)
ax4.set_title(f'S100',  fontsize=22);

fig.suptitle(f'Unifrom distribution:\nHistogram of S1 S2 S10 and S100',  fontsize=25);
fig.tight_layout()
fig.subplots_adjust(top=0.8,hspace = 0.5,wspace = 0.4)
plt.show()
```

### Exponential dice (i.e. exponential distribution)

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(17, 12))

ax1.hist(Data_Exp_S1,bins=100,range=(-1,2), color="yellow",density=True)
ax1.set_xlabel('X[exp]',fontsize=20)
ax1.set_ylabel('NUMBERS',fontsize=20)
ax1.grid(True)
ax1.set_title(f'S1',  fontsize=22);

ax2.hist(Data_Exp_S2,bins=100,range=(-1,2), color="gold",density=True)
ax2.set_xlabel('S2[exp]',fontsize=20)
ax2.set_ylabel('NUMBERS',fontsize=20)
ax2.grid(True)
ax2.set_title(f'S2',  fontsize=22);

ax3.hist(Data_Exp_S10,bins=100,range=(-1,2), color="orange",density=True)
ax3.set_xlabel('S10[exp]',fontsize=20)
ax3.set_ylabel('NUMBERS',fontsize=20)
ax3.grid(True)
ax3.set_title(f'S10',  fontsize=22);

Bins=np.linspace(0.,2.,100)
gauss=np.array([gaussian(x,1,np.sqrt(1./100)) for x in Bins])
ax4.hist(Data_Exp_S100,bins=100,range=(0,2), color="red",density=True,label="numerical")
ax4.plot(Bins,gauss, color="blue",label="theorical")
ax4.set_xlabel('S100[exp]',fontsize=20)
ax4.set_ylabel('NUMBERS',fontsize=20)
ax4.grid(True)
ax4.legend(fontsize=20)
ax4.set_title(f'S100',  fontsize=22);

fig.suptitle(f'Exponential distribution:\nHistogram of S1 S2 S10 and S100',  fontsize=25);
fig.tight_layout()
fig.subplots_adjust(top=0.8,hspace = 0.5,wspace = 0.4)
plt.show()
```

### Cauchy dice (i.e. cauchy distribution)

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(17, 12))

ax1.hist(Data_Cauchy_S1,bins=100,range=(-3,3), color="lightgray",density=True)
ax1.set_xlabel('X[exp]',fontsize=20)
ax1.set_ylabel('NUMBERS',fontsize=20)
ax1.grid(True)
ax1.set_title(f'S1',  fontsize=22);

ax2.hist(Data_Cauchy_S2,bins=100,range=(-3,3), color="silver",density=True)
ax2.set_xlabel('S2[exp]',fontsize=20)
ax2.set_ylabel('NUMBERS',fontsize=20)
ax2.grid(True)
ax2.set_title(f'S2',  fontsize=22);

ax3.hist(Data_Cauchy_S10,bins=100,range=(-3,3), color="gray",density=True)
ax3.set_xlabel('S10[exp]',fontsize=20)
ax3.set_ylabel('NUMBERS',fontsize=20)
ax3.grid(True)
ax3.set_title(f'S10',  fontsize=22);

Bins=np.linspace(-3,3.,100)
Cauchy=np.array([cauchy(x,0,1) for x in Bins])
ax4.hist(Data_Cauchy_S100,bins=100,range=(-3,3), color="black",label="numerical",density=True)
ax4.plot(Bins,Cauchy, color="red",label="theorical")
ax4.set_xlabel('S100[exp]',fontsize=20)
ax4.set_ylabel('NUMBERS',fontsize=20)
ax4.grid(True)
ax4.legend(fontsize=20)
ax4.set_title(f'S100',  fontsize=22);

fig.suptitle(f'Cauchy distribution:\nHistogram of S1 S2 S10 and S100',  fontsize=25);
fig.tight_layout()
fig.subplots_adjust(top=0.8,hspace = 0.5,wspace = 0.4)
plt.show()
```

As asserts the central limit theorem, the analysis dysplayed upon show that for the first two distributions the gaussian is the limit of the $S_i$. 
On the other hand, for the Cauchy distribuction the limit of the $S_i$ is again another Cauchy distribution.
