#### First part

In the **first part** of the excercise must be calculated the mean value of the uniform distribution in the interval $[0,1]$: In other words the following integral $\langle r \rangle = \int_0^1 r dr = 1/2$. 
The intergal has been calculated with the method of the average.

The final estimation, indeed, has been calculated with the blocking method that consist in dividing the set of measurements in blocks, and considering as the best estimation the mean of the means of the data in each block. Then comes naturaly consider the estimations of the errors as the standard deviation of the means of the blocks. For this part of the excercise has been used a total of $10^6$ estimation divided in 1000 blocks of 1000 estimations and have been pinted the final estimations as functios of the blocks.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
M=1000000
N=1000
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
L=int(M/N)
x = np.arange(N) 


sum_prog = np.loadtxt(fname = "valori_medie.txt")
err_prog = np.loadtxt(fname = "errori_medie.txt")
      
      

x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,sum_prog-0.5,yerr=err_prog,c="blue")
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```


As it can be seen from the graph, after an inital transient phase where the average fluctuates around 0.5, the expected value converges. Morever the interval, fixed by the final averges value and its error, contains the theorical prediction 1/2.

#### Second Part

The **second part** of the excercise consists in valuating the variation $\sigma^2$ of the uniform distribution in the interval $[0,1]$. The method, implemented to reach the scope, is not different from the first part, but of course, the integral has been changed: $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$.

To calculate the estimation has been used the blocking method again: has been used a total of $10^6$ estimation divided in 1000 blocks of 1000 estimations.

```python
M=1000000
N=1000
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
L=int(M/N)
x = np.arange(N) 


sum_prog = np.loadtxt(fname = "valori_varianze.txt")
err_prog = np.loadtxt(fname = "errori_varianze.txt")
      
      

plt.errorbar(x,sum_prog-1/12,yerr=err_prog,c="blue")
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```

As the graph of the average, after an inital transient phase where the mean of the variance fluctuates around 1/12, the expected value converges and the estimation is confident with the analytical value.

#### Third Part

The scope of the **third part** of the excercise is verify the effectiviness of the pseudo-random number generator. To reach this goal have been implemented 100 $\chi^2$ tests: one for every $10^4$ values sampled to calculate the estimations in the previous cells. 

To implement the $\chi^2$ test, the interval $[0,1]$ has been divided in 100 sub-intervals. Every sub-interval must be filled by the PRN Generator $n_s$ times, whose expected value is $n_t=100$ and whose $\sigma=10$.

$$
\chi^2_i = \sum_{j=1}^{M=100} \chi^2_ {ij} = \sum_{j=1}^{M=100} \frac{(n_{s j}-n_t )^2}{\sigma}
$$

```python
dati= np.loadtxt(fname = "dati.txt")
```

```python
M=100
L=int(1000000/M)
n_t=L/M
chi=np.zeros(M)
x=np.zeros(M)

for i in range(M):#values on the x axis
    x[i]=i+1

for i in range(M):#in this cycle chi^2 are filled
    sper=np.zeros(L)#data used for the i-th chi^2 test
    n_s=np.zeros(M)
    for j in range(L):#at every cycle data are shifted of L=10000
        prog=int(i*L)+j
        sper[j]=dati[prog]
    
    for j in range(M):#the values that are in the sub-intervals are counted
        for k in range(L):
            if (1/M*(j)<sper[k]<1/M*(j+1)):
                n_s[j]=n_s[j]+1
            
    
    for j in range(M):#calculation of the chi^2
        chi[i]+=((n_s[j]-n_t)**2)/n_t 
```

```python
print(f' mean founded: {np.mean(chi)}')
plt.scatter(x,chi,s=10,c="black")
plt.plot(x,chi,c="blue")
plt.xlabel('n')

plt.ylabel(f'$\chi^{2}$')
plt.grid(True)
plt.title(f'100 values of $\chi^{2}$\neach for different set of data of size $10^4$')
plt.show() 
```

Although most of the values sit around the mean value=100, from the plot we can see that a lot of values are distant from it for an amount that exceeds $\sigma$. Sometimes more than $3 \sigma$: this should be very unlikely.

Below has been incremented the number of sub-intervals to check if something changes.

```python
print(f' mean founded: {np.mean(chi)}')
plt.scatter(x,chi,s=10,c="black")
plt.plot(x,chi,c="blue")
plt.xlabel('n')

plt.ylabel(f'$\chi^{2}$')
plt.grid(True)
plt.title(f'100 values of $\chi^{2}$\neach for different set of data of size $10^4$')
plt.show() 
```
