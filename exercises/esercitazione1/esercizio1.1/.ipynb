```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
M=100000
N=100
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
L=int(M/N)
x = np.arange(N) 


sum_prog = np.loadtxt(fname = "valori_medie.txt")
err_prog = np.loadtxt(fname = "errori.txt")
      
      

x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,sum_prog-0.5,yerr=err_prog)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```
