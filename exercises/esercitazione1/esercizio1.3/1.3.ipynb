### EXERCISE 01.3

The exercise 1.3 is a simulation of the Buffon experiment.
A needle of length $L$ is thrown at random onto a horizontal plane ruled with straight lines at distance $d$ (must be $d > L$, but not $d\gg L$ otherwise $P\ll 1$). The probability $P$ that the needle intersects one of these lines is: $P = 2L/\pi d$. This could be used to evaluate $\pi$ from throws of the needle: if the needle is thrown down $N_{thr}$ times and is observed to land on a line $N_{hit}$ of those times, we can make an estimate of $\pi$ from
$$
\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}
$$

As usual has been shown a picture of the estimations of $\pi$ and their uncertainties as a function of the number of blocks.

For the simulation have been arranged 100 blocks, in which have been done 100 experiments and, for every experiment, the needle has been thrown $10^3$ times.

#### How the experiment is implemented

Recognizing that the problem is translation invariant, to simplify the structure of the experiment, the space has been implemented just with 2 parallel lines distant $D=2$.The needle is built as a segment of length $L=1.95$. Subsequently a random number $Y_{cm}$ is extracted uniformly between $[0,2]$ and rapresents the position touched by the midpoint of the needle. Than, is extracted the angle $\theta$ between the prolungation of the needle and the parallel lines uniformly between $[0,\pi]$. The condition for which one of the 2 lines is hitted by the needle is indeed:
$$
(Y_{cm} + \frac{L}{2}sin(\theta) > D) || (Y_{cm} - \frac{L}{2}sin(\theta) < 0)
$$


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math as m
```

```python
sum_prog = np.loadtxt(fname = "Datasum_prog.txt")
err_prog = np.loadtxt(fname = "Dataerr_prog.txt")
```

```python
N=100
x = np.arange(N) 
x*=100000
      
plt.errorbar(x,sum_prog-m.pi,yerr=err_prog,c="blue",label="numerical")
plt.plot(x,np.zeros(x.size),c="red",label="theorical")
plt.xlabel('#throws')
plt.ylabel('<pi>-pi')
plt.grid(True)
plt.legend(loc=(0.7,0.2))
plt.show()
```

As usual the value converges as the number of blocks increases and the errors decrease. The final estimation (calculated with 100 blocks) is consistent with the exact value.
