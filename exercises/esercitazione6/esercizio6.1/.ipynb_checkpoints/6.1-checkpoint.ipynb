### EXERCISE 06.1

Has been completed the C++ code to simulate 1D Ising model with 
- the Metropolis sampling algorithm and
- the Gibbs sampling algorithm

to sample the Boltzmann's weight $p(\{ s_i \}) = \exp(-\beta H)/Z$,
adding also the possibility to restart from a previous spin configuration.

Has been added the calculation of $C(N,T)$, $\chi(N,T)$, $M(N,T,h)$ and has been computed:
1. $U(N,T) = \langle H \rangle_{h=0}$
2. $C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ 
3. $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$
4. $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$

as a function of $T$ in the range $\left[0.5;2.0\right]$ for a system of $N=50$ spins with $J=1$.

As usual has been used data blocking and has been given an estimate of the statistical uncertainties.

```python
import numpy as np
import matplotlib.pyplot as plt
points=100
J = 1.0
Ns = 50
```

#### Ising model

The Ising Model is a kind of lattice models, which are a big industry in statistical mechanic. In particular the Ising model is useful to describe magnetic system. It is known from quantum mechanic that the magnetic moment is proportional to the spin of the electrons that has two degrees of freedom: $+1/2$ or $-1/2$. Indeed it is convenient to represent a magnetic system as a lattice, whose cells have 2 degrees of freedom. Moreover the lattice must be transational invariant (possible with the implementation of periodic boundary condition) and the contributions to the energy come from the coupling of every spin to an external magnetic field and from the coupling of every spin with its neighbours.

In this excercise it has been simulated an Ising model in one dimention, whose hamiltonian can be written as follows
$$
H = -J\sum_{i=1}^{50} s_i s_{i+1}-\frac{h}{2}\sum_{i=1}^{50} (s_i + s_{i+1})
$$

where $J$ is the costant of the energy coupling and can take only poitive values (beacuse the system was supposed to be ferromagnetic) and it was fixed equal to 1 for simplicity, $h$ is the external magnetic field, and $s_i$ can take two values: +1 or -1.

#### Time evolution

In the simulation below it has been started from a random configuration of 50 spins and, at fixed temperature, the system has been evolved with 2 different algorithm: Metropolis and Gibbs, in the canonical ensamble.

When the evolution is ruled by Metropolis algorithm every step montercarlo consists in a suggestion of a swapping of every spin in the configuration and the proposal can be accepted or rejected.
When the evolution is ruled by Gibbs algorithm, the calculator look at the neighbours of every spin and subsequently calculate a probability of setting the spin up and (consequently) a probability of setting the spin down. The spin is then placed following this partiction. 

Obviously, in both algorithm, are favored configurations with large boltzmann weight that correspond to configurations with low energy.

#### Measurements

After a sufficient equilibration time, needed due to the random configuration from which the evolution start, the system can be considered at the equilibrium of a particular temperature (that has been fixed in the "input.dat" file) and measurements can be taken. To implement the equilibration, as it has been done in exercise 4, it has been created a variable **m_BOOL** that
- when is false the configuration is created randomly
- when is true the system start from the last configuration of the last simulation. 

For every quantity the estimations have been made with the average blocking method and the errors have been calculated as the standard deviation of the mean. Compared to the values of the estimations the errors are too small to be seen in the graph. Indeed, to overcome this problem, it has been plotted a graph of differences between the numerical values and the analytical values.

### Analysis of Internal Energy U/N

Values of the Internal Energy per particles and comparison with analytic results.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))

T = np.linspace(0.2,3.0,num=points)
beta = 1/T
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )


# plot theoretical result
ax1.plot(T, e,
    color="red",
    lw=1,
    label="Theory")

numerical_gibbs_e = np.loadtxt(fname = "Ene_gibbs_graph.out",usecols=0)
err_gibbs_e = np.loadtxt(fname = "Ene_gibbs_graph.out",usecols=1)

numerical_metro_e = np.loadtxt(fname = "Ene_metropolis_graph.out",usecols=0)
err_metro_e = np.loadtxt(fname = "Ene_metropolis_graph.out",usecols=1)

numerical_T=np.arange(0.5,2.1,0.1)

# plot numerical result
ax1.errorbar(numerical_T, numerical_gibbs_e,yerr=err_gibbs_e,
          color="blue",
          lw=1,
          label="Gibbs")

ax1.errorbar(numerical_T, numerical_metro_e,yerr=err_metro_e,
          color="green",
          lw=1,
          label="Metropolis")


# add labels for axes
ax1.set_xlabel("Temperature",fontsize=20)
ax1.tick_params(axis="x", labelsize=15)
ax1.tick_params(axis="y", labelsize=15)
ax1.grid(True)
ax1.set_ylabel('U/N',fontsize=20)

# add legend
leg = ax1.legend(loc=(0.75, 0.2),frameon=True)
leg.get_frame().set_edgecolor('black')
leg.get_frame().set_linewidth(3)


# add a title (e.g. that says what dimension we used)
ax1.set_title("Internal Energy Theory vs Numerics:\nGibbs and Metropolis ")





th = np.tanh(J/numerical_T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

# plot numerical result
ax2.errorbar(numerical_T, numerical_gibbs_e-e,yerr=err_gibbs_e,
          color="blue",
          lw=1,
          label="Gibbs")

ax2.errorbar(numerical_T, numerical_metro_e-e,yerr=err_metro_e,
          color="green",
          lw=1,
          label="Metropolis")


# add labels for axes
ax2.set_xlabel("Temperature ",fontsize=20)
ax2.tick_params(axis="x", labelsize=15)
ax2.tick_params(axis="y", labelsize=15)
ax2.hlines(0, 0.5, 2.1, label='Theory')
ax2.grid(True)

# add legend
leg = ax2.legend(loc=(0.95, 0.7),frameon=True)
leg.get_frame().set_edgecolor('black')
leg.get_frame().set_linewidth(3)


# add a title (e.g. that says what dimension we used)
ax2.set_title("Internal Energy Theory - Numerics:\nGibbs and Metropolis ")



```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))
T = np.linspace(0.2,3.0,num=points)
beta = 1/T

h=0.02 #external field
b = 1/T
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)



# plot theoretical result
ax1.plot(T, M,
    color="red",
    lw=1,
    label="Theory")

numerical_gibbs_M = np.loadtxt(fname = "Mag_gibbs_graph.out",usecols=0)
err_gibbs_M = np.loadtxt(fname = "Mag_gibbs_graph.out",usecols=1)

numerical_metro_M = np.loadtxt(fname = "Mag_metropolis_graph.out",usecols=0)
err_metro_M = np.loadtxt(fname = "Mag_metropolis_graph.out",usecols=1)

numerical_T=np.arange(0.5,2.1,0.1)

# plot numerical result
ax1.errorbar(numerical_T, numerical_gibbs_M,yerr=err_gibbs_M,
          color="blue",
          lw=1,
          label="Gibbs")

ax1.errorbar(numerical_T, numerical_metro_M,yerr=err_metro_M,
          color="green",
          lw=1,
          label="Metropolis")


# add labels for axes
ax1.set_xlabel("Temperature ",fontsize=20)
ax1.tick_params(axis="x", labelsize=15)
ax1.tick_params(axis="y", labelsize=15)
ax1.set_ylabel('magnetization M',fontsize=20)
ax1.grid(True)
# add legend
leg = ax1.legend(loc=(0.75, 0.7),frameon=True)
leg.get_frame().set_edgecolor('black')
leg.get_frame().set_linewidth(3)


# add a title (e.g. that says what dimension we used)
ax1.set_title("magnetization M with h = 0.02 Theory vs Numerics:\nGibbs and Metropolis ")


beta = 1/numerical_T

h=0.02 #external field
b = 1/numerical_T
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)




# plot numerical result
ax2.errorbar(numerical_T, numerical_gibbs_M-M,yerr=err_gibbs_M,
          color="blue",
          lw=1,
          label="Gibbs")

ax2.errorbar(numerical_T, numerical_metro_M-M,yerr=err_metro_M,
          color="green",
          lw=1,
          label="Metropolis")


# add labels for axes
ax2.set_xlabel("Temperature",fontsize=20)
ax2.tick_params(axis="x", labelsize=15)
ax2.tick_params(axis="y", labelsize=15)
ax2.hlines(0, 0.5, 2.1, label='Theory')
ax2.grid(True)

# add legend
leg = ax2.legend(loc=(0.65, 0.2),frameon=True)
leg.get_frame().set_edgecolor('black')
leg.get_frame().set_linewidth(3)


# add a title (e.g. that says what dimension we used)
ax2.set_title("magnetization M with h = 0.02 Theory - Numerics:\nGibbs and Metropolis ")
```

##### An observation

In the simulation made, it can be seen in every quantity measured that, at low temperature, the uncertaincies increase. What does it happen?

By way of example has been reported in the pyton cells below the data measured of the magnetization at $T=0.5$ and at $T=2.0$. At lower temperature the metropolis accepts less moves and the efficiency decreases. Indeed it can be seen, when $T=2.0$, that the data are more distributed around the mean value. This causes bigger fluctuations in the average of the blocks at low temperature and, indeed, a bigger error, which is defined as the standard deviation of the mean of the blocks.
