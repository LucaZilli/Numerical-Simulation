### EXERCISE 03.1 

Black-Scholes theory assumes that the time evolution of the asset price is stochastic and exhibits geometric Brownian motion (GBM) with constant risk-free interest rate $r$ and volatility $\sigma$.

In this exercitation the scope is simulate the trend of the price of an asset and give an estimations of the price of the **call** option and **put** option on that asset.To give the correct estimation are given the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$


During the theory lesson has been seen that if it has been supposed that the asset's price behave like a geometric brownian motion, the evolution at the time $T$ can be estimated like
$$
S(T) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)t+\sigma W(T)\right]=S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)t+\sigma Z\sqrt{T}\right]
$$
where $W(T)$ is stochastic variable that follows the distribution $\mathcal{N}(0, T)$, while $Z$ is a stochastic variable that follows the distribution $\mathcal{N}(0, 1)$. Indeed are obtained the estimations of the **call** and the **put** option like $C=e^{-r T}\max[0, S(T)-K]$ e $P=e^{-r T}\max[0, K-S(T)]$.


Have been done 100 blocks of 1000 simulation each and, for each simulation has been calculated $C$ and $P$ as explained upon. Subsequently has been plotted the estimations using data blocking techinque.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Here below the python cell that contain the analytic result of the call-option and put-option price at the delivery time T.

```python
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

In the following cells I will show you the simulation made. 

I divided my work in 100 blocks of 1000 simulations each. Every block contain an estimation of the price at the delivery time T. Indeed I must calculate the Averages of the 100 estimations obtained as function of blocks and the error is calculated as the variance of the averages.

### Direct sampling:

```python
Call_N = np.loadtxt(fname = "C_N_prog.txt")
err_Call = np.loadtxt(fname = "err_C_prog.txt")

Put_N = np.loadtxt(fname = "P_N_prog.txt")
err_Put = np.loadtxt(fname = "err_P_prog.txt")
```

```python

M=100000
N=100

L=int(M/N)
x = np.arange(N) 
x*=L # Number of estimations
plt.errorbar(x,Call_N-call,yerr=err_Call, color="red")
plt.xlabel('#estimations')
plt.ylabel('<call>-call')
plt.grid(True)
plt.show()


plt.errorbar(x,Put_N-put,yerr=err_Put, color="blue")
plt.xlabel('#estimations')
plt.ylabel('<put>-put')
plt.grid(True)
plt.show()

```



It has been compared the simulation made with the analytical value plotting the difference between the estimation and the value itslef. What it can be inferred is that both the Put and Call options converge and that the estimations with its respective errors are consistent with the expected result.

### Discretized sampling:

The second part of the exercise is almost identycal and has the same goal of the first one. The main difference is in the evolution of the Asset price. With the disretized sampling the interval $[0,T]$ must be divided in 100 sub-intervals and iterating the ricorsive formula
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$ where $Z_i$ is a random variable extracted from $\mathcal N (0, 1)$.

```python
Call_N_discreto = np.loadtxt(fname = "C_N_discreto_prog.txt")
err_Call_discreto = np.loadtxt(fname = "err_C_discreto_prog.txt")

Put_N_discreto = np.loadtxt(fname = "P_N_discreto_prog.txt")
err_Put_discreto = np.loadtxt(fname = "err_P_discreto_prog.txt")
      
```

```python

M=100000
N=100

L=int(M/N)
x = np.arange(N) 

Call_N_discreto = np.loadtxt(fname = "C_N_discreto_prog.txt")
err_Call_discreto = np.loadtxt(fname = "err_C_discreto_prog.txt")

Put_N_discreto = np.loadtxt(fname = "P_N_discreto_prog.txt")
err_Put_discreto = np.loadtxt(fname = "err_P_discreto_prog.txt")
      
      

x*=L # Number of estimations


plt.errorbar(x,Call_N_discreto-call,yerr=err_Call_discreto,color="red")
plt.xlabel('# estimations')
plt.ylabel('<call_discreto>-call')
plt.grid(True)
plt.show()

plt.errorbar(x,Put_N_discreto-put,yerr=err_Put_discreto,color="blue")
plt.xlabel('# estimations')
plt.ylabel('<put_discreto>-put')
plt.grid(True)
plt.show()
```
