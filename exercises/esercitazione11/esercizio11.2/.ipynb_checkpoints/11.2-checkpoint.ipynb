### EXCERCISE 11.2

In this excercise was extended the model to obtain a fit of the following polynomial of order 3:

$$
f(x)=4-3x-2x^2+3x^3
$$
for $x \in [-1,1]$.

In order to make practice with NN, has been explored different choices for:

- the number of layers
- the number of neurons in each layer
- the activation function
- the optimizer
- the loss function
  
After having observed and summarized which Neural Network performs better, have been tested a a fit on values out of the interval on which the NN has been trained on.

```python
# target parameters
a = 4 
b = -3
c = -2
d = 3
```

```python
import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 10000)
x_valid = np.random.uniform(-1, 1, 1000)
x_valid.sort()
y_target = a + b*x_valid + c*x_valid**2 + d*x_valid**3 # ideal (target) linear function

sigma = 0.5 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(a + b*x_train + c*x_train**2 + d*x_train**3, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(a + b*x_valid + c*x_valid**2 + d*x_valid**3, sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

### Structure of the neural network

From the analysis made, the conclusion achieved are that the form of the NN does not bring substuntial differences in the output if the number of parameters of the neural networks are kept equal (or similar). But has been seen a problem in the oscillations of the loss function. With a network built with 9 layers and 46 neurons arranged like 1 2 4 8 16 8 4 2 1, has been seen a more ingent number of oscilations than if it is used a network arranged like 20 15 10 1.

#### Network arranged like 1 2 4 8 16 8 4 2 1

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
from tensorflow.keras import optimizers, losses, metrics

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))
model.add(Dense(2, input_shape=(1,),activation='relu'))
model.add(Dense(4, input_shape=(1,),activation='relu'))
model.add(Dense(8, input_shape=(1,),activation='relu'))
model.add(Dense(16, input_shape=(1,),activation='relu'))
model.add(Dense(8, input_shape=(1,),activation='relu'))
model.add(Dense(4, input_shape=(1,),activation='relu'))
model.add(Dense(2, input_shape=(1,),activation='relu'))
model.add(Dense(1, input_shape=(1,),activation='relu'))
# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
#model.compile(optimizer='sgd', loss=losses.categorical_crossentropy, metrics=['mse'])
from tensorflow.keras import optimizers, losses, metrics

# Configure the model for mean-squared error regression.
#model.compile(optimizer=optimizers.SGD(),             #optimizer='sgd',       # stochastic gradient descent
              #loss=losses.categorical_crossentropy,
              #metrics=[metrics.categorical_accuracy]) #metrics=['accuracy'])  # an optional list of metrics
```

```python
# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=50,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
for l in model.layers:
    w = l.get_weights()
    print("layer nodes weights: ", w[0].shape)
    print("layer bias weights: ", w[1].shape)
```

```python
#model.evaluate(x_valid, y_valid, batch_size=32)
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(['train', 'test/validation'], loc='best')
plt.show()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
x_predicted = np.random.uniform(-1, 1, 1000)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```
