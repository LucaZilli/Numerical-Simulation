On the other hand it has been reached the scope following another way: using the "Path Integral Ground State" algorithm.

#### How does the PIGS algorithm work?

Given an hamiltonian H, whose solution is ignote and a state $|\Psi_T \rangle$, supposed to be the state in which the system is , it has been seen during theory lessons that the state evolve like $|\Psi_T(\tau) \rangle=e^{-\tau\hat{H}} |\Psi_T \rangle$ where $\tau=i\frac{t}{\hbar}$. If the state is described in a complet system of eigensate of the hamiltonian, since by definition the eigenvelue of the GS is the lowest, in the limit of "big" $\tau$, is true that $|\Psi_T(\tau) \rangle \sim a_0 e^{-\tau E_0}$ where $a_0$ is the product bra-ket between the groundstate $|\Psi_0 \rangle$ and $|\Psi_T \rangle$. This implies 
$
|\Psi_0 \rangle = \lim\limits_{\tau \to \infty} e^{-\tau \hat{H}} |\Psi_T \rangle
$.
Introducing M times the identity, the ground state wavefunction can be written like 
$$
\Psi_0(x) \cong \int dx_1 ... dx_M G\left(x, x_1, \frac{\tau}{M}\right) ... G\left(x_{M-1}, x_M, \frac{\tau}{M}\right) \Psi_T (x_M)
$$
where $G\left(x_i, x_j, \frac{\tau}{M}\right)=\langle x_i | e^{-\frac{\tau}{M} \hat{H}} | x_j \rangle$ is the Feynman propagator.

Now is important to find a way to write the expetation value of an observable $\hat{A}$ on the ground state that is possible to sample with an algorithm. During theory lessons it was shown that can be written like

$$
\langle \hat{A} \rangle_0
\cong \int \prod\limits_{i=1}^{2M+1} dx_i \hat{A}(x_{M+1}) \frac{\Psi_T(x_1) \prod\limits_{j=1}^{2M}G\left(x_j, x_{j+1}, \frac{\tau}{M}\right)\Psi_T(x_{2M+1})}{\langle \Psi_T(\tau) | \Psi_T(\tau) \rangle}
$$
and now is easy to see that the second factor inside the integral is a distribution probability that can be sampled.

On the propagators is applied the primitive approsimation where are neglected terms $O\left(\left(\frac{\tau}{M}\right)^2\right)$ in the esponent. Indeed the propagator can be written like 
$$
\langle x_i | e^{-\frac{\tau}{M} \hat{H}} | x_j \rangle \cong e^{-\frac{\tau}{2M} V(x_i)} \langle x_i | e^{-\frac{\tau}{M} \hat{T}} x_j \rangle e^{-\frac{\tau}{2M} V(x_j)}
$$
and the propagator of the kinetic energy can be easily calculated as follows:
$$
\langle x_i | e^{-\frac{\tau}{M} \hat{T}} | x_j \rangle = \frac{e^{-\frac{|x_i-x_j|^2}{4\gamma \tau / M}}}{(4\pi \gamma \tau /M)^{3/2}}
$$
Written like this, the propagator, can be seen like a Boltzamn wheigh of an armonic oscillator. The immaginary temporal evolution now can be seen like a system of classical particles interacting with rmonical potential: a **polymer**, that indeed, names the algorithm.

What it can be seen is that the ground state founded with this algorithm approximate clearly better the ground state of the matrix problem and that the convergence of $|\Psi_T \rangle$ is obviously faster if the algorithm begin the evolution with a wavefunction that looks similar to the final one.

### Matrix problem

```python
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
#N = 1000 # number of iterations
N = 400 # number of iterations



# Step sizes
#x = np.linspace(-a/2, a/2, N)
x = np.linspace(-4.9875, 4.9875, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])
print("1st excited state energy: ", E[1])
print("2nd excited state energy: ", E[2])
```

# ANALYSIS OF PIMC ALGORITHM

#### How does change the path integral method when $T>0$?

It is known from quantum statistical mechanic that, in the canonical ensamble (i.e. at fixed temperature), the average value of an observable A is 
$
\langle \hat{A} \rangle = \frac{\int dx \rho(x, x, \beta) A(x)}{Tr(\hat{\rho})}
$
where $\hat{\rho}(x, x', \beta)=\langle x | e^{-\beta\hat{H}} | x' \rangle $ is called matrix density.

If are introduced M identity in the mean value is obtained
$$
\langle \hat{A} \rangle = \int dx_1 ... dx_M A(x_1) \frac{\rho\left(x_1, x_2, \frac{\beta}{M}\right)...\rho\left(x_M, x_1, \frac{\beta}{M}\right)}{Tr(\hat{\rho})}
$$

and for the matrix density is applied the primitive approsimation
$$
\rho\left(x_i, x_j, \frac{\beta}{M}\right)\cong e^{-\frac{\beta}{2M}V(x_i)} \frac{e^{-\frac{M|x_i-x_j|^2}{4\lambda \beta}}}{(4\pi \lambda \beta/M)^{3/2}} e^{-\frac{\beta}{2M}V(x_j)}
$$

An other time the system of a quantum single particle can be seen as a polymer of M classical particles interacting with an armonic potential.
