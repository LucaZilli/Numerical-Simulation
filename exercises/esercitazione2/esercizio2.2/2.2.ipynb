### EXERCISE 02.2


In this exercise has been simulated a random walk of 100 steps in two different ways:
1. on a cubic lattice with lattice constant $a=1$; at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$
2. in the continuum; at each discrete time the walker makes a step of length equal to $a(=1)$ along a **random direction** obtained by sampling **uniformly** the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$

The goal of the simulation is to verify the trend
$
\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}=k\sqrt{N}
$ of the distance after $N$ steps of a diffusive system.

The simulation has been divided in 100 blocks of 100 measurements of ${\langle |\vec{r}_N|^2 \rangle_{RW}}$ each, for a total of $10^4$ random walks. As usual the estimations are calculated as the averages of the blocks and has been plotted $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ as function of the steps with all the blocks taken on account.

###  Random walk on a lattice

Why a random walk follows the law 
$
\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}=k\sqrt{N}
$? To answer this question it has been demonstrated using induction method.

Induction method (RW in 1D), may $L$ be equal to $+1$ or $-1$ with equal probability, such that ${r}_n=\sum^N_0{L}$:

1. Base of induction N=1:
$${\langle |{r}_1|^2 \rangle}= 1^2 0.5 + (-1)^2 0.5 = 1$$

2. inductive step (the relation is true for N-1):
$${\langle |{r}_N|^2 \rangle}= {\langle |{r}_{N-1} + L|^2 \rangle}= {\langle |{r}_{N-1}|^2 \rangle} + {\langle |L|^2 \rangle} +  2 {\langle ({r}_{N-1}  L) \rangle}={\langle |{r}_{N-1}|^2 \rangle} + {\langle |L|^2 \rangle}= N - 1 + 1 = N $$

where ${\langle ({r}_{N-1}  L) \rangle}= -0.5 {r}_{N-1} + 0.5  {r}_{N-1}= 0$

This implies that $\sqrt{\langle |{r}_N|^2 \rangle_{RW}}=k\sqrt{N}$ with $k=1$.

### Continuos random walk

Even a continuos RW follows the trend 
$
\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}=k\sqrt{N}
$. To demonstrate this, has been used an induction method.

Induction method (RW in 2D), may $\vec{L}$ be isotropic such as $|\vec{L}|=1$ and $\vec{r}_n=\sum^N_0\vec{L}$:

1. Base of induction N=1:
$${\langle |\vec{r}_1|^2 \rangle}= {\langle \vec{r}_1*{r}_1 \rangle} = |\vec{L}|^2=1$$

2. inductive step (the relation is true for N-1):
$${\langle |\vec{r}_N|^2 \rangle}= {\langle |\vec{r}_{N-1} + L|^2 \rangle}= {\langle |\vec{r}_{N-1}|^2 \rangle} + {\langle |L|^2 \rangle} +  2 {\langle (\vec{r}_{N-1}  L) \rangle}={\langle |\vec{r}_{N-1}|^2 \rangle} + {\langle |L|^2 \rangle}= N - 1 + 1 = N $$

where ${\langle (\vec{r}_{N-1} * \vec{L}) \rangle} \sim  \langle cos(\theta) \rangle = 0$

This implies that $\sqrt{\langle |{r}_N|^2 \rangle_{RW}}=k\sqrt{N}$ with $k=1$.
