### EXERCISE 02.1

In this exercise has been calculated numericaly the integral written below
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

To reach the scope, the integral has been calculated with the montecarlo method of the average, following two different paths:

1. sampling a uniform distribution in $[0,1]$
2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

Than has been plotted the esimations of $I$ and their uncertainties as function of the number of blocks.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

#### Unform distribution:

To solve the problem, in the first case, have been calculated 100 estimations of the integral and each of them calculated using 1000 extractions, indeed 
$$
{\langle I \rangle}_{i} = \frac{1}{N}\sum_{k=1}^{N=1000} f(x_k) = \frac{1}{N}\sum_{k=1}^{N=1000} \frac{\pi}{2}\cos(\pi x_j/2)
$$ 
with $x_j$ sorted randomly in $[0, 1]$.
Subsequently has been calculated the 100 estimations obtained as follows 
$$\langle I \rangle = \frac{1}{M} \sum_{i=1}^{M=100} \langle I \rangle_i
$$ 
and their uncertencies as the standard deviation of the mean.

```python
N=100
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
x = np.zeros(N) 

for i in range(N):
    x[i]=i


sum_prog = np.loadtxt(fname = "valori_integrali.txt")
err_prog = np.loadtxt(fname = "errori_integrali.txt")
print(f'last error is {err_prog[-1]}')        

#x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,sum_prog-1,yerr=err_prog)
plt.xlabel('#blocks')
plt.ylabel('<I>-1')
plt.grid(True)
plt.show()
```

#### Important sampling:

Instead of sampling an uniform distribution in the interval $[0,1]$, it is possible extract, in the interval we are interested in, points that follow a distribution $p(x)$ and the same integral it can be written as follows:
$$
\int f(x) dx = \int \frac{f(x)}{p(x)} p(x)
$$

It can be proofed that the precision can increase if the function $p(x)$ follows a similar trend of the integrand function $f(x)$.
To extract numbers that follow a probability distribution $p(x)$ has been implemented the method of the inverse of the cumulative.

Estabilished that, to solve the integral have been calculated 100 estimations of it and each of them calculated using 1000 extractions, indeed 
$$
{\langle I \rangle}_{i} = \frac{1}{N}\sum_{k=1}^{N=1000} f(x_k) = \frac{1}{N}\sum_{k=1}^{N=1000} \frac{\pi}{2}\cos(\pi x_k/2)
$$ 
with $x_k$ sorted following the distribution $p(x)=2-2x$ in $[0, 1]$.
Subsequently has been calculated the 100 estimations obtained as follows 
$$\langle I \rangle = \frac{1}{M} \sum_{i=1}^{M=100} \langle I \rangle_i
$$ 
and their uncertencies as the standard deviation of the mean.

In the following cell is shown $f(x)$ and $p(x)$ used on the simulation in the interval of interest $[0,1]$. 

```python
bins=np.linspace(start=0, stop=1, num=1001)
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(bins,np.cos(np.pi/2.*bins)*np.pi/2.,c="blue",label="Integrand")
ax.plot(bins,2-2*bins,c="red",label="P. Distribution")
ax.grid(True)
ax.set_title("Intergrand and Probability Distribuction")
ax.legend(loc=(0.77, 0.73))
```

```python
N=100
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
x = np.zeros(N) 

for i in range(N):
    x[i]=i


sum_prog = np.loadtxt(fname = "valori_integrali_imp.txt")
err_prog = np.loadtxt(fname = "errori_integrali_imp.txt")
print(f'last error is {err_prog[-1]}')  
      

#x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,sum_prog-1,yerr=err_prog)
plt.xlabel('#blocks')
plt.ylabel('<I>-1')
plt.grid(True)
plt.show()
```


What it can be seen from the analyis made is that both the methods used are effective to the evaluation of the integral and both the best estimations are consistent with the analytic solution.

Moreover the integral estimated with important sampling reveals a smallest error on the estimation, as expected from the theory.
