```python
import numpy as np
import matplotlib.pyplot as plt
```

Uploading data from the simulation:

```python
epot_solid = np.loadtxt(fname = "output.epot.solid",usecols=1)
pres_solid = np.loadtxt(fname = "output.pres.solid",usecols=1)

epot_liquid = np.loadtxt(fname = "output.epot.liquid",usecols=1)
pres_liquid = np.loadtxt(fname = "output.pres.liquid",usecols=1)

epot_gas = np.loadtxt(fname = "output.epot.gas.0",usecols=1)
pres_gas = np.loadtxt(fname = "output.pres.gas.0",usecols=1)

print(epot_solid)
print(pres_solid)
```

Just few functions i will need to calculate standard deviations and autocorrelation:

```python
import math

def error2(data,N):  # Function for statistical uncertainty estimation
    if N==0:
        return 0
    else:
        ave=0.0
        av2=0.0
        for i in range (N):
            ave += data[i]
            av2 += data[i]*data[i]
        ave=ave/N
        av2=av2/N
            
        return av2 - ave**2
    
def err(data,N):
    err=np.zeros(N)
    for i in range(N):
        err[i] = math.sqrt( error2(data,i) )
    return err
```

```python
def AutoCorrelation(data,N):
    chi=np.zeros(N)
    sigma=error2(data,N)
    for t in range(N):
        t_max=N-t
        ave_1=0.0
        ave_2=0.0
        av2=0.0
        for t_primo in range(t_max):
            ave_1 += data[t_primo]
            ave_2 += data[t_primo+t]
            av2 += data[t_primo]*data[t_primo+t]
    
        ave_1 = ave_1/t_max
        ave_2 = ave_2/t_max
        av2 = av2/t_max
    
        value = ( av2 - ave_1*ave_2 )/sigma
        chi[t] = value
    
    return chi
    
```

## SOLID PHASE:

```python
"""Analysis of data"""
M=2000
t=np.arange(M)
error_epot=err(epot_solid,M)
chi_epot=AutoCorrelation(epot_solid,M)
error_pres=err(pres_solid,M)
chi_pres=AutoCorrelation(pres_solid,M)



"""Plots and layout"""
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(17, 12))

"""Potential Enrgy"""
ax1.plot(t,chi_epot,color='blue')
ax1.set_title(f'Autocorrelation of Potential Energy',  fontsize=22);
# add labels for axes
ax1.set_xlabel("# Measures",fontsize=20)
ax1.set_ylabel("Autocorrelation",fontsize=20)
ax1.tick_params(axis="x", labelsize=15)
ax1.tick_params(axis="y", labelsize=15)
ax1.grid(True)

ax2.plot(t,error_epot,color='red')
ax2.set_title(f'Standard Deviation of Potential Energy',  fontsize=20);
# add labels for axes
ax2.set_xlabel("# Measures ",fontsize=20)
ax2.set_ylabel("dev std",fontsize=20)
ax2.tick_params(axis="x", labelsize=15)
ax2.tick_params(axis="y", labelsize=15)
ax2.grid(True)


"""Pression"""
ax3.plot(t,chi_pres,color='green')
ax3.set_title(f'Autocorrelation of Pression',  fontsize=22);
# add labels for axes
ax3.set_xlabel("# Measures",fontsize=20)
ax3.set_ylabel("Autocorrelation",fontsize=20)
ax3.tick_params(axis="x", labelsize=15)
ax3.tick_params(axis="y", labelsize=15)
ax3.grid(True)

ax4.plot(t,error_pres,color='orange')
ax4.set_title(f'Standard Deviation of Pression',  fontsize=20);
# add labels for axes
ax4.set_xlabel("# Measures ",fontsize=20)
ax4.set_ylabel("dev std",fontsize=20)
ax4.tick_params(axis="x", labelsize=15)
ax4.tick_params(axis="y", labelsize=15)
ax4.grid(True)

fig.suptitle(f'Potential Energy and Pression of the $Solid$ Phase:\nAutocorrelation and Standard Deviation after {M} steps',  fontsize=25);
fig.tight_layout()
fig.subplots_adjust(top=0.8,hspace = 0.5,wspace = 0.4)
plt.show()
```
