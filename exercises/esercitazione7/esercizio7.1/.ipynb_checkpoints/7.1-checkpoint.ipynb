### EXERCISE 07.1


In this exercise is simulated a system of many particles interacting to each other with a potential of Lennard-Jones for every phase of matter, as happens in excercise 4. The systems start in a configuration of FCC and is brought to a configuration chosen by the input parameters after a transient equilibration phase. The parameters, for every phase of matter, are:
- solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
- liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
- gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

This time, on the other hand, the temporal evolution is not integrated from the differential equation as happens with Verlet algorithm, but is ruled by the metropolis algorithm, whose transitional probability is controlled by an input parameter. This parameter has been changed in order to obtain an acceptance close to 50%. The probability that must be maximized is the Boltzamnn distribution, in order obtain a system with less energy as possible with the temperature that has been fixed from the "input.dat" file.

Once the system is equilibrated, the internal energy per particle and the pressure are calculated with the blocking methods' averages and their uncertaincies are estimated as the standard deviations of the means of the blocks.

In this particular jupyter notebook the goal is to understand which must be the size of the blocks needed to calculate the estimations. Indeed the autocorrelation of the measurements and the standard deviations of the blocks' averages are calculated as the number of steps in each blocks vary. In particular, the standard deviation, is an increasing function that saturates for a cartain number of steps. In the cells below it was tried to disocover the minimum number of steps to avoid autocorrelations between measurements of different block and an underestimation of the errors.


```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

Uploading data from the simulation:

```python
epot_solid = np.loadtxt(fname = "data_Epot_solid.txt",usecols=1)
pres_solid = np.loadtxt(fname = "data_Pres_solid.txt",usecols=1)


epot_liquid = np.loadtxt(fname = "data_Epot_liquid.txt",usecols=1)
pres_liquid = np.loadtxt(fname = "data_Pres_liquid.txt",usecols=1)

epot_gas = np.loadtxt(fname = "data_Epot_gas.txt",usecols=1)
pres_gas = np.loadtxt(fname = "data_Pres_gas.txt",usecols=1)
```

Just few functions i will need to calculate standard deviations and autocorrelation:

```python
def error2(data,N):  # Function for statistical uncertainty estimation
    if N==0:
        return 0
    else:
        ave=0.0
        av2=0.0
        for i in range (N):
            ave += data[i]
            av2 += data[i]*data[i]
        ave=ave/N
        av2=av2/N
            
        return av2 - ave**2
```

```python
def error(AV,AV2,n):  
    if n==0:
        return 0
    else:
        return math.sqrt((AV2 - AV**2)/n)
    
def err(data,M,L_list):
    err=[]
    for L in L_list:                 
        N=int(M/L)            
        ave = np.zeros(N)         
        av2 = np.zeros(N)     
        sum_prog = 0     
        su2_prog = 0       
        err_prog = 0                       
                           
        for i in range(N):
            sum = 0
            for j in range(L):
                k = j+i*L
                sum += data[k]
            ave[i] = sum/L       
            av2[i] = (ave[i])**2 

        for i in range(N):
    
            sum_prog += ave[i] 
            su2_prog += av2[i] 
        sum_prog/=N 
        su2_prog/=N 
        err_prog = error(sum_prog,su2_prog,N) 
        err.append(err_prog)
    return err
        
```

```python
def AutoCorrelation(data,N):
    chi=np.zeros(N)
    sigma=error2(data,N)
    for t in range(N):
        t_max=N-t
        ave_1=0.0
        ave_2=0.0
        av2=0.0
        for t_primo in range(t_max):
            ave_1 += data[t_primo]
            ave_2 += data[t_primo+t]
            av2 += data[t_primo]*data[t_primo+t]
    
        ave_1 = ave_1/t_max
        ave_2 = ave_2/t_max
        av2 = av2/t_max
    
        value = ( av2 - ave_1*ave_2 )/sigma
        chi[t] = value
    
    return chi
    
```


From the analysis made it can be seen that for every phase of the system the errors, estiamte as the standard deviation of the means of the blocks, can be considered saturated for a number of steps larger than 1000. For the same number of steps even the data are uncorrelated.


### SOLID PHASE:

calulation of the autocorrelation and errors as function of steps in each block for the solid phase.

```python
"""Analysis of data"""
M=500000
L_list=[10,20,50,100,125,400,500,800,1000,1500,2000,2500,3125,4000,5000]
range_chi=8000
t_chi=np.arange(range_chi)
t=np.arange(M)
error_epot=err(epot_solid,M,L_list)
chi_epot=AutoCorrelation(epot_solid,range_chi)
error_pres=err(pres_solid,M,L_list)
chi_pres=AutoCorrelation(pres_solid,range_chi)
```

```python

"""Plots and layout"""
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(17, 12))

"""Potential Enrgy"""
ax1.plot(t_chi,chi_epot,color='blue')
ax1.set_title(f'Autocorrelation of Potential Energy',  fontsize=22);
# add labels for axes
ax1.set_xlabel("# Measures",fontsize=20)
ax1.set_ylabel("Autocorrelation",fontsize=20)
ax1.tick_params(axis="x", labelsize=15)
ax1.tick_params(axis="y", labelsize=15)
ax1.grid(True)

ax2.plot(L_list,error_epot,color='red')
ax2.set_title(f'Standard Deviation of Potential Energy',  fontsize=20);
# add labels for axes
ax2.set_xlabel("# steps per block ",fontsize=20)
ax2.set_ylabel("dev std",fontsize=20)
ax2.tick_params(axis="x", labelsize=15)
ax2.tick_params(axis="y", labelsize=15)
ax2.grid(True)


"""Pression"""
ax3.plot(t_chi,chi_pres,color='green')
ax3.set_title(f'Autocorrelation of Pression',  fontsize=22);
# add labels for axes
ax3.set_xlabel("# Measures",fontsize=20)
ax3.set_ylabel("Autocorrelation",fontsize=20)
ax3.tick_params(axis="x", labelsize=15)
ax3.tick_params(axis="y", labelsize=15)
ax3.grid(True)

ax4.plot(L_list,error_pres,color='orange')
ax4.set_title(f'Standard Deviation of Pression',  fontsize=20);
# add labels for axes
ax4.set_xlabel("# steps per block ",fontsize=20)
ax4.set_ylabel("dev std",fontsize=20)
ax4.tick_params(axis="x", labelsize=15)
ax4.tick_params(axis="y", labelsize=15)
ax4.grid(True)

fig.suptitle(f'Potential Energy and Pression of the $Solid$ Phase:\nAutocorrelation and Standard Deviation after {M} steps',  fontsize=25);
fig.tight_layout()
fig.subplots_adjust(top=0.8,hspace = 0.5,wspace = 0.4)
plt.show()
```
