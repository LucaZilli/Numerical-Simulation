### EXERCISE 10.2
In this exercise has been modified the code of the excercise 9 to be carried on by four processors in parallel way.

Each processor, in the rappresentation used, represent a continent, in which a population composed by 100 organisms evolves indipendently.
After 50 generation, the best organism of every continent migrates to another one and the implementation of the order of the migration is sorted randomly every time.

The conseguence of the modification of the algoithm in this way is a gain in efficency.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_cmap(n, name='hsv'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)
```

```python
BestPath_rank0 = np.loadtxt(fname = "BestPath_rank0.txt")
BestPath_Length2_rank0 = np.loadtxt(fname = "BestPath_Length2_rank0.txt")
BestHalf_Length2_rank0 = np.loadtxt(fname = "BestHalf_Length2_rank0.txt")

BestPath_rank1 = np.loadtxt(fname = "BestPath_rank1.txt")
BestPath_Length2_rank1 = np.loadtxt(fname = "BestPath_Length2_rank1.txt")
BestHalf_Length2_rank1 = np.loadtxt(fname = "BestHalf_Length2_rank1.txt")

BestPath_rank2 = np.loadtxt(fname = "BestPath_rank2.txt")
BestPath_Length2_rank2 = np.loadtxt(fname = "BestPath_Length2_rank2.txt")
BestHalf_Length2_rank2 = np.loadtxt(fname = "BestHalf_Length2_rank2.txt")

BestPath_rank3 = np.loadtxt(fname = "BestPath_rank3.txt")
BestPath_Length2_rank3 = np.loadtxt(fname = "BestPath_Length2_rank3.txt")
BestHalf_Length2_rank3 = np.loadtxt(fname = "BestHalf_Length2_rank3.txt")
```

### Best Path for each continent:

```python
x_coordinates_square = np.loadtxt(fname = "coordinates_square.txt",usecols=0)
y_coordinates_square = np.loadtxt(fname = "coordinates_square.txt",usecols=1)

x_coordinates_square = np.loadtxt(fname = "coordinates_square.txt",usecols=0)
y_coordinates_square = np.loadtxt(fname = "coordinates_square.txt",usecols=1)
```

```python
N=32 #Number of cities
x = np.arange(N) 
path_0=np.zeros((N,2))
path_1=np.zeros((N,2))
path_2=np.zeros((N,2))
path_3=np.zeros((N,2))
for i in range (N):
    path_0[i][0] = x_coordinates_square[ int(BestPath_rank0[i]-1) ]
    path_0[i][1] = y_coordinates_square[ int(BestPath_rank0[i]-1) ]
    path_1[i][0] = x_coordinates_square[ int(BestPath_rank1[i]-1) ]
    path_1[i][1] = y_coordinates_square[ int(BestPath_rank1[i]-1) ]
    path_2[i][0] = x_coordinates_square[ int(BestPath_rank2[i]-1) ]
    path_2[i][1] = y_coordinates_square[ int(BestPath_rank2[i]-1) ]
    path_3[i][0] = x_coordinates_square[ int(BestPath_rank3[i]-1) ]
    path_3[i][1] = y_coordinates_square[ int(BestPath_rank3[i]-1) ]
  
    
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(17, 17))

ax1.scatter(path_0[0][0], path_0[0][1], marker = 'x',color="black",s=50)
ax1.plot(path_0[:,0],path_0[:,1],color="black",lw=1)
ax1.grid(True)
ax1.set_ylabel(f'y',fontsize=20)
ax1.set_xlabel(f'x',fontsize=20)
ax1.tick_params(axis="x", labelsize=15)
ax1.tick_params(axis="y", labelsize=15)
ax1.set_title(f'Best Path:\nContinent 0',  fontsize=20);

ax2.scatter(path_1[0][0], path_1[0][1], marker = 'x',color="black",s=50)
ax2.plot(path_1[:,0],path_1[:,1],color="black",lw=1)
ax2.grid(True)
ax2.set_ylabel(f'y',fontsize=20)
ax2.set_xlabel(f'x',fontsize=20)
ax2.tick_params(axis="x", labelsize=15)
ax2.tick_params(axis="y", labelsize=15)
ax2.set_title(f'Best Path:\nContinent 1',  fontsize=20);

ax3.scatter(path_2[0][0], path_2[0][1], marker = 'x',color="black",s=50)
ax3.plot(path_2[:,0],path_2[:,1],color="black",lw=1)
ax3.grid(True)
ax3.set_ylabel(f'y',fontsize=20)
ax3.set_xlabel(f'x',fontsize=20)
ax3.tick_params(axis="x", labelsize=15)
ax3.tick_params(axis="y", labelsize=15)
ax3.set_title(f'Best Path:\nContinent 2',  fontsize=20);

ax4.scatter(path_3[0][0], path_3[0][1], marker = 'x',color="black",s=50)
ax4.plot(path_3[:,0],path_3[:,1],color="black",lw=1)
ax4.grid(True)
ax4.set_ylabel(f'y',fontsize=20)
ax4.set_xlabel(f'x',fontsize=20)
ax4.tick_params(axis="x", labelsize=15)
ax4.tick_params(axis="y", labelsize=15)
ax4.set_title(f'Best Path:\nContinent 3',  fontsize=20);


cmap = get_cmap(len(path_0))
for i in range(N):
    ax1.scatter(path_0[i][0], path_0[i][1], color=cmap(i))
    ax2.scatter(path_1[i][0], path_1[i][1], color=cmap(i))
    ax3.scatter(path_2[i][0], path_2[i][1], color=cmap(i))
    ax4.scatter(path_3[i][0], path_3[i][1], color=cmap(i))

fig.suptitle(f'Best Path for each continent',  fontsize=25);
fig.tight_layout()
fig.subplots_adjust(top=0.9,hspace = 0.3,wspace = 0.4)
plt.show()
```

One conseguence of the indipendent evolution for each continent, is that the pace of convergence of a continent can be different from an other one.
It has been deliberately done an evolution that can show this characteristic: all the continents converges to the best algorithm apart the continent 1, whose path has the cities 23-24 changed in the order.

On the other hand, to solve the problem of the salesman traveler is sufficient that just one continent converges to the best possible path. Indeed it can be satisfied of the results of this simulation.

### $L^2$ Norm of the Best path for each continent:


```python
M=600 #Number of generations
L=50  #Number steps before a print
N=int(M/L) #Number of prints
t = np.arange(N)*L

"""Plots and layout"""
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(17, 12))

"""Continent 0"""
ax1.plot(t,BestPath_Length2_rank0,color='blue')
ax1.set_title(f'Continent 0: $L^2$ Best path',  fontsize=22);
ax1.set_xlabel("# generations",fontsize=20)
ax1.set_ylabel(f'$L^2$ ',fontsize=20)
ax1.tick_params(axis="x", labelsize=15)
ax1.tick_params(axis="y", labelsize=15)
ax1.grid(True)
print(BestPath_Length2_rank0[9])
"""Continent 1"""
ax2.plot(t,BestPath_Length2_rank1,color='red')
ax2.set_title(f'Continent 1: $L^2$ Best path',  fontsize=22);
ax2.set_xlabel("# generations",fontsize=20)
ax2.set_ylabel(f'$L^2$ ',fontsize=20)
ax2.tick_params(axis="x", labelsize=15)
ax2.tick_params(axis="y", labelsize=15)
ax2.grid(True)

"""Continent 2"""
ax3.plot(t,BestPath_Length2_rank2,color='green')
ax3.set_title(f'Continent 2: $L^2$ Best path',  fontsize=22);
ax3.set_xlabel("# generations",fontsize=20)
ax3.set_ylabel(f'$L^2$ ',fontsize=20)
ax3.tick_params(axis="x", labelsize=15)
ax3.tick_params(axis="y", labelsize=15)
ax3.grid(True)

"""Continent 3"""
ax4.plot(t,BestPath_Length2_rank3,color='orange')
ax4.set_title(f'Continent 3: $L^2$ Best path',  fontsize=22);
ax4.set_xlabel("# generations",fontsize=20)
ax4.set_ylabel(f'$L^2$ ',fontsize=20)
ax4.tick_params(axis="x", labelsize=15)
ax4.tick_params(axis="y", labelsize=15)
ax4.grid(True)


fig.suptitle(f'$L^2$ norm of the legth of the best path\nfor each continent (alias rank)',  fontsize=25);
fig.tight_layout()
fig.subplots_adjust(top=0.8,hspace = 0.5,wspace = 0.4)
plt.show()
```

From the graph above, at the first glance, it can be thought that the algotithm converges in less than 400 steps, but is not true. The last drop of the graph is difficult to be seen due to a small configuration of the path similar in length to the shortest. 
It has been already told that the pace of the convergence is different for every block. In this simulation the fastest continents to reach the shortest configuration have been the continent 2 and the continent 3 and have been taken 450 generations to converge.

In the excercise 09.1, for the same problem and for the cities placed in the same coordinates, has been taken 2300 generations to reach the convergence.
**More than 5 times the number of generations**.

In the graph shown upon it can be seen the trend of the fitness of the best half of the population vs the number of generations.
